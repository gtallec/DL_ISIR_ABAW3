```python
def get_toy_task_matrix_from_crossval(experiment_folder, n_fold, n_task):
    experiments = os.listdir(experiment_folder)
    task_matrices = np.zeros((n_fold, n_task, n_task))
    for i in range(n_fold):
        experiment = os.path.join(experiment_folder, 'fold_{}.json'.format(i))
        with open(experiment) as json_file:
            json_dict = json.load(json_file)
            visual_experiment_parser = VisualExperimentParser(json_dict)
        log_folder = visual_experiment_parser.get_log_folder()
        model = visual_experiment_parser.get_model()
        model.build((None, 299, 299, 3))
        model_weights = os.path.join(log_folder, 'weights', 'ckpt')
        model.load_weights(model_weights, 'all')
        task_matrices[i, :, :] = model.built_models['regressor']['instance'].get_task_matrix()
    return task_matrices
        
def get_toy_task_matrix(experiment_folder, n_task):
    experiments = os.listdir(experiment_folder)
    n = len(experiments)
    task_matrices = np.zeros((n, n_task, n_task))
    for i in range(len(experiments)):
        experiment = experiments[i]
        with open(os.path.join(experiment_folder, experiment)) as json_file:
            json_dict = json.load(json_file)
            log_folder = json_dict['log_folder']
            model_dict = json_dict['model']
            visual_experiment_parser = VisualExperimentParser(json_dict)
        log_folder = visual_experiment_parser.get_log_folder()
        model = visual_experiment_parser.get_model()
        model_weights = os.path.join(log_folder, 'weights', 'ckpt')
        model.load_weights(model_weights, 'all')
        task_matrices[i, :, :] = model.built_models['regressor']['instance'].get_task_matrix()
    return task_matrices
    
```

```python
# Computing pairwise mutual information matrix
pairwise_mutual_info_mat = np.zeros((8, 8))
for i in range(8):
    for j in range(8):
        pairwise_mutual_info_mat[i, j] = reducers.paiwise_mutual_information(i, j)(dataset)
```

```python
inverse_entropies = np.diag(np.diag(1/pairwise_mutual_info_mat))
pairwise_mutual_info_mat_normalized = ((inverse_entropies) @ pairwise_mutual_info_mat @ (inverse_entropies))
mask_diag = np.ones_like(inverse_entropies) - np.eye(inverse_entropies.shape[0]) + np.diag(np.diag(pairwise_mutual_info_mat))
print(mask_diag)
sns.heatmap(pairwise_mutual_info_mat_normalized * mask_diag)
    
```

```python
import itertools
combinations = itertools.combinations([0, 1, 2, 3, 4, 5, 6, 7], 4)

def complementary(a, els):
    return [el for el in els if el not in a]


def partition(a, k):
    L = []
    combinations = itertools.combinations(a, k)
    for combination in combinations:
        list_combination = list(combination)
        L.append((list_combination, complementary(list_combination, a)))
    return L

def convert_to_AU(index_labels):
    return [labels[index_label] for index_label in index_labels]

def best_partition_disfa(fold, n_au=8, batch_size=32, T_sample=1, k=4):
    disfa_labels = (gen_disfa(fold=fold, n_au=n_au, batch_size=batch_size, T_sample=T_sample)
                    .map(get_label, tf.data.experimental.AUTOTUNE)
                    .prefetch(tf.data.experimental.AUTOTUNE)
                    .cache())
    partitions = partition(range(n_au), k)
    mutual_informations = [(general_pairwise_mutual_information(disfa, combi, compl_combi)/np.sqrt(entropy(disfa, combi) * entropy(disfa, compl_combi))).numpy() 
                           for (combi, compl_combi) in partitions]
    best_partition_index = np.argmin(mutual_informations)
    first_cluster, second_cluster = partitions[best_partition_index]
    return (convert_to_AU(first_cluster), convert_to_AU(second_cluster))    
```
