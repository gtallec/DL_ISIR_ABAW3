# General Tools : 

We define the classical binary entropy as :
$$ H(p) = - (p\log{(p)} + (1 - p)\log{(1 -p)})$$

```python
def H(p):
    return -p * np.log(p) - (1 - p) * np.log(1 - p)

p = np.linspace(0.01, 0.99, 50)

plt.plot(p, H(p))
plt.xlabel("p")
plt.ylabel("H(p)")
plt.title("Binary Entropy")
plt.show()
```

We also define a slight twist of Bernouilli with parameter $p$ that we denote $\mathcal{B}(p)$, such that the distribution of $Y \sim \mathcal{B}(p)$ verifies :
$$p_Y = (1 - p)\delta_{-1} + p \delta_{1}$$

## 1. First Toy

Say we want an $M$ task toy : 
$Y \sim \mathcal{B}(q), \forall n\in [1, M] : Y_n \sim Y \mathcal{B}(p_n)$.
We also define $q_k = p_{Y_k}(1) = qp_k + (1-q)(1-p_k).$




$$H(Y_n \mid Y_{-n}) - H(Y_k \mid Y_{-k}) = g(p_n) - g(p_k)$$

```python
def g(p, q):
    return H(q * p + (1 - q) * (1 - p))

def dp_g(p, q):
    return (2*q - 1) * np.log(((1 - q) * p + q * (1 - p))/(q * p + (1 - q) * (1 - p)))

p = np.linspace(0.01, 0.99, 50)
q = np.linspace(0.01, 0.99, 50)
P, Q = np.meshgrid(p, q)

fig= plt.figure()
ax = fig.add_subplot(111, projection='3d')
ax.set_xlabel("p")
ax.set_ylabel("q")
ax.set_zlabel("g(p, q)")
ax.plot_wireframe(P, Q, g(P, Q)) 
plt.show()

    
```

```python
# prendre q loin de 1/2 et p_1, p_2 d'entropie différente.
q = 0.001
p_n = np.linspace(0.01, 0.99, 50)
p_k = np.linspace(0.01, 0.99, 50)

P_n, P_k = np.meshgrid(p_n, p_k)
fig= plt.figure()
ax = fig.add_subplot(111, projection='3d')
ax.set_xlabel("p_n")
ax.set_ylabel("p_k")
ax.set_zlabel("g(p_n, 1e-3) - g(p_k, 1e-3)")
ax.plot_wireframe(P_n, P_k , g(P_n, q) - g(P_k, q)) 
plt.show()
```
