```python
import tensorflow as tf
import matplotlib.pyplot as plt
import os

from tensorboard.backend.event_processing.event_accumulator import EventAccumulator
```

```python
def numpy_from_tf(path):
        tf_size_guidance = {
        'compressedHistograms': 10,
        'images': 0,
        'scalars': 1000,
        'histograms': 1,
        'tensors': 1000
    }
    
        event_acc = EventAccumulator(path, tf_size_guidance)
        event_acc.Reload()
        
        return event_acc
    
def decode(val):
    tensor_proto = val.tensor_proto
    return tf.make_ndarray(tensor_proto)
    
```

```python
path = os.path.join('..', 'logs', 'toy', 'vertical_split', 'sanity', '1_best', 'metrics', 'frob_to_answer', 'train')
real_path = os.path.join(path, os.listdir(path)[0])
print(real_path)


event_acc = numpy_from_tf(real_path)
frob_stream = event_acc.Tensors('frob')

frob_stream = [decode(step) for step in frob_stream]
print(frob_stream)
```
